```python
import pandas as pd
import numpy as np
```

```python
gdsc_df = pd.read_csv("datasets/gdsc_members.csv", header=None, names=['First Name', 'Last Name', 'Email', 'Company', 'Title',
 'Registered At', 'Events Registered Counts'])
```

```python
gdsc_df.head()
```

```python
# missing values in GDSC dataset
gdsc_df.isnull()
```

```python
# check missing values per column
gdsc_df.isnull().any()
```

```python
# replace null(falsy) values with "member"
gdsc_df["Title"].fillna("Member", inplace=True)
gdsc_df["Company"].replace(np.nan, "UR-CST", inplace=True)
```

```python
gdsc_df.head()
```

```python
# check types of fields
gdsc_df.dtypes
```

```python
gdsc_df['Registered At'] = pd.to_datetime(gdsc_df['Registered At'])
```

```python
gdsc_df.head()
```

```python
# change date format of "registered at"
gdsc_df['Registered At'] = gdsc_df['Registered At'].dt.strftime('%Y-%m-%d')
```
